```python
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv("/content/drive/MyDrive/Kaggle Contest/1/train.csv")
```

```python
train_data.head()
```

```python
women = train_data.loc[train_data.Sex == 'female']["Survived"]
rate_women = sum(women)/len(women)
rate_women
print("% of women who survived:", rate_women)
```

```python
men = train_data.loc[train_data.Sex == 'male']["Survived"]
rate_men = sum(men)/len(men)

print("% of men who survived:", rate_men)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()

train_data["Gender"] = lb.fit_transform(train_data["Sex"])
train_data
```

```python
y = train_data["Survived"]
```

```python
X = train_data[["Pclass", "Gender", "SibSp", "Parch"]]
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
test_data = pd.read_csv("/content/drive/MyDrive/Kaggle Contest/1/test.csv")
test_data.head()
```

```python
lb = LabelEncoder()

test_data["Gender"] = lb.fit_transform(test_data["Sex"])
test_data
```

```python
predictions = model.predict(test_data[["Pclass", "Gender", "SibSp", "Parch"]])
predictions
```

```python
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
